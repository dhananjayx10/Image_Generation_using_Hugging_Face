```python
import requests
import io
from PIL import Image
from IPython.display import display

API_URL = "https://api-inference.huggingface.co/models/stabilityai/stable-diffusion-xl-base-1.0"
headers = {"Authorization": "Bearer YOUR TOKEN"} # ISSE ACCHA NEXT MODEL USE KAR SAKTE HO 

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.content

image_bytes = query({
    "inputs": "an impossibly deep sinkhole with crystal clear water at the bottom, a mermaid with red accents in the middle of the sinkhole on a rock. some stone rocks making a pathway leads to a staircase on the side of the sinkhole",
})

# Convert the image bytes to a PIL image
image = Image.open(io.BytesIO(image_bytes))

# Display the image
display(image)
```

```python
# Use huggingface model black-forest-labs/FLUX.1-dev
```

```python
import aiohttp
import asyncio

async def query(data):
    async with aiohttp.ClientSession() as session:
        async with session.post(
            "https://api-inference.huggingface.co/models/black-forest-labs/FLUX.1-dev",
            headers={
                "Authorization": "Bearer YOUR TOKEN",  # GOOD MODEL I EVER FOUND FROM DHANANJAY
                "Content-Type": "application/json"
            },
            json=data
        ) as response:
            content_type = response.headers.get("content-type")
            if "application/json" in content_type:
                result = await response.json()
                print(result)
            else:
                blob = await response.read()
                with open("output.png", "wb") as f:
                    f.write(blob)
                print(" Image saved as output.png")

# Jupyter / Colab me direct await use karo
await query({"inputs": "an impossibly deep sinkhole with crystal clear water at the bottom, a mermaid with red accents in the middle of the sinkhole on a rock. some stone rocks making a pathway leads to a staircase on the side of the sinkhole"})
```
